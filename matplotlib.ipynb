```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
```

```python
#  scatter plot, also called a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram, is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y, c = 'y', alpha = .7, marker ='s')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title('x vs y')
plt.grid()
plt.figure(figsize=(6,2))
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create the subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, first plot
plt.plot(x, y1, label="Line 1", color="blue")
plt.title("Line 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Second subplot
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, second plot
plt.plot(x, y2, label="Line 2", color="green")
plt.title("Line 2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Third subplot
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, third plot
plt.plot(x, y3, label="Line 3", color="red")
plt.title("Line 3")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Fourth subplot
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, fourth plot
plt.plot(x, y4, label="Line 4", color="purple")
plt.title("Line 4")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"], patch_artist=True, 
            boxprops=dict(facecolor="skyblue", color="blue"),
            medianprops=dict(color="red"))
```
